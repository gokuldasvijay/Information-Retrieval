```python
#doc_path=input("Give the path of document:")
import string
def Document(*doc_path):
  doc_list=[]
  for p in doc_path:
    df=''
    df=open(p,"r")
    i=df.read()
    #for punctuation removal
    punct = '''!()-[]{};:'"\,<>./?@#$%^&*_~'''
    for ele in i:
      if ele in punct:
        i= i.replace(ele, "")
    #tokenization
    doc_list.append(i.split())
    
  return doc_list
```

```python
def tokenise(doc_list):
  tokens=[]
  for y in doc_list:
    tokens.append(y.split())
  return tokens
```

```python
def stopwords(tokens):
  from nltk.corpus import stopwords
  nltk.download('stopwords')
  stop_words = stopwords.words('english')
  txt=[]
  output=[]
  for li in tokens:
    for i in range(0,len(li)):
      word=li[i]
      word=word.lower()
      if word not in stop_words:
            output.append(li[i])
            #txt = (" ").join(output)
  return output
```

```python
import string
def punctuation(tokens):
  pnkt=[]
  for t in tokens:
    if t not in string.punctuation:
      pnkt.append(t)
  return pnkt
```

```python
def unique_words(pnkt):
  tx=[]
  for t in pnkt:
    if t not in tx:
      tx.append(t)

  return(tx)   
```

```python
def indexing(w_list,index,q):
#q should be the output list of Document() function
#w_list should be the output list of unique_words() function
  for wd in w_list:
    for w in range(0,len(q)):
      if wd in q[w]:
        index[w].append(1)
      else:
        index[w].append(0)
  return index
```

```python
def Dictionary(w_list,vec_list):
  wlst=[]
  ulst=[]
  for wd in w_list:
    li=[]
    wlst.append(wd)
    for w in range(0,len(vec_list)):
      if wd in vec_list[w]:
        li.append(1)
      else:
        li.append(0)
        ulst.append(li)

  wd_vec_dct=dict(zip(wlst,ulst)) 
  
  return(wd_vec_dct)

```
