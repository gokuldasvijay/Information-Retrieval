```python
import string
def Document(*doc_path):
  doc_list=[]
  for p in doc_path:
    df=''
    df=open(p,"r")
    i=df.read()
    #for punctuation removal
    punct = '''!()-[]{};:'"\,<>./?@#$%^&*_~'''
    for ele in i:
      if ele in punct:
        i= i.replace(ele, "")
    #tokenization
    doc_list.append(i.split())
    
  return doc_list
```

```python
def stopwords(tokens):
  from nltk.corpus import stopwords
  nltk.download('stopwords')
  stop_words = stopwords.words('english')
  txt=[]
  output=[]
  for li in tokens:
    for i in range(0,len(li)):
      word=li[i]
      word=word.lower()
      if word not in stop_words and word.isalpha():
            output.append(li[i])
            
  return output
```

```python
def Dictionary(w_list,vec_list):
  wlst=[]
  ulst=[]
  for wd in w_list:
    li=[]
    wlst.append(wd)
    for w in range(0,len(vec_list)):
      if wd in vec_list[w]:
        li.append(str(w+1))
      else:
        break
      ulst.append(li)

  wd_vec_dct=dict(zip(wlst,ulst)) 
  
  return(wd_vec_dct)
```
